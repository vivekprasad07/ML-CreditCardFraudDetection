```python
#Split the data into training data and testing data

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape, x_test.shape)
```

```python
#Model training
#Using Logistic regression for binary Classification problem 0 1

model = LogisticRegression()
```

```python
#training the Logistics regression Model with Training Data

model.fit(x_train,y_train)
```

```python
#Model Evaluation based on accuracy scoree

#Accuracy Score
#accuracy on trainig data 
x_train_prediction = model.predict(x_train)
trainig_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on trainign data : ', trainig_data_accuracy)
```

```python
#accuracy on test data 

x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction,y_test)
```
